```python
import numpy as np
import random as rdm
import matplotlib.pyplot as plt
from scipy import stats
```

```python
popul = np.random.normal(100, 15, 100000)  # popul is an array that contains 100,000 values

# now let's plot the distribution
bins = np.linspace(50, 150, 30)
plt.hist(popul, bins)
plt.tick_params(labelsize=16)
plt.xlabel("Individual IQ's", size=24)
plt.ylabel('Frequency', size=24)
plt.xlim(40, 160)
plt.ylim(0, 10000);
```
